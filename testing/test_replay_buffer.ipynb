## storeTransition

Overwriting the oldest entries takes place.

```python
for i in range(0, 5):
    phi = torch.randn(frameShape)
    action = torch.randint(high = 4, size = (1,)) # must be a tuple
    reward = torch.randint(high = 100, size = (1,))
    nextState = torch.randn(frameShape)
    transition = (phi, action, reward, nextState)
    foo.storeTransition(transition)

    print(f"---\n{foo.index = }\nReward:\t{foo.reward}")
```
