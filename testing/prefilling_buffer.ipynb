```python
FRAME_HEIGHT = 256
FRAME_WIDTH = 240
BUFFER_SIZE = 10

torch.manual_seed(5004 + 96);
rb = ReplayBuffer(BUFFER_SIZE, (FRAME_HEIGHT, FRAME_WIDTH) )

# overfill to prove that there is no failure from doing so
for i in range(BUFFER_SIZE * 2):
    rb.storeTransition((
        torch.randn( (FRAME_HEIGHT, FRAME_WIDTH) ),
        torch.randint(high = 4, size = (1,)),
        torch.randint(high = 100, size = (1,)),
        torch.randn( (FRAME_HEIGHT, FRAME_WIDTH) )
    ))
```
