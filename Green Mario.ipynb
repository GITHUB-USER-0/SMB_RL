```python
# Q-learning, World 1-1, Simple and V3
Q_agent = QLearningAgent("simple", world=1, stage=1, rom_version='v3', seed=5004)
Rewards_q = Q_agent.train(episodes=30)
```

```python
# DQN, World 1-1, Simple, V3
DQN_agent = TorchDQNAgent(
    actionSpace=SIMPLE_MOVEMENT,
    world=1,
    stage=1,
)

Rewards_DQN = DQN_agent.train(episodes=30)
```

```python
plt.plot(Rewards_q, label='Q-Learning')
plt.plot(Rewards_DQN, label='DQN')
plt.axhline(np.mean(Rewards_q), label='Q-Learning average', linestyle='--', color='red')
plt.axhline(np.mean(Rewards_DQN), label='DQN average', linestyle='--', color='green')
plt.xlabel('Episode')
plt.ylabel('Total Reward')
plt.title('Simple Movement on 30 Episodes for World 1-1')
plt.legend()
plt.show()
```
