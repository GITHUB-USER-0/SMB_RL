# Comments on different course features

## Warp zones

* [1-2](https://www.mariowiki.com/World_1-2_(Super_Mario_Bros.))
* [4-2](https://www.mariowiki.com/World_4-2_(Super_Mario_Bros.))

These levels introduces a **warp zone**, that allows the player to skip multiple worlds within the game. If the agent were incentivized to complete the *game* quickly, this would presumably be highly rewarding.

At present, the reward function is simpler and focused on moving right, not running out the clock, and not dying.

The current level of cropping of the top of the screen *may* present a further barrier to accessing the warp zone.

## Puzzle levels

* [4-4](https://www.mariowiki.com/World_4-4_(Super_Mario_Bros.))
* [7-4](https://www.mariowiki.com/World_7-4_(Super_Mario_Bros.))
* [8-4](https://www.mariowiki.com/World_8-4_(Super_Mario_Bros.))

In these puzzle levels the player is presented with branching paths. If the correct path is chosen, the player can progress. Otherwise, the player will repeat some portions of the level.

This presents an interesting challenge for an agent. It seems unlikely that a DQN approach would be able to learn this behavior, and given the default reward algorithm, it is plausible that the agent would in fact perversely learn to take the wrong path because that would conceivably allow for consistent reward for x velocity, relative to completing the course.

If the agent were able to navigate this level without any challenge, it may also indicate that the model has become overfit as it has learned the specific path to take.

## Underwater levels

* [2-2](https://www.mariowiki.com/World_2-2_(Super_Mario_Bros.))
* [7-2](https://www.mariowiki.com/World_7-2_(Super_Mario_Bros.))
  
There are two courses that are almost entirely underwater.

In other courses there are underwater segments, but not the full level, eg., [5-2](https://www.mariowiki.com/World_5-2_(Super_Mario_Bros.)), [6-2](https://www.mariowiki.com/World_6-2_(Super_Mario_Bros.)).

This meaningfully changes how Mario interacts, as he goes from walking/running and jumping to swimming.

In early testing, performance on the water level appeared reasonably high.

## Specific levels

### [5-3](https://www.mariowiki.com/World_5-3_(Super_Mario_Bros.))

This level is an interesting challenge as it may strongly incentivize 'NOOP' inputs as the player waits on a platform for the platform itself to move.
