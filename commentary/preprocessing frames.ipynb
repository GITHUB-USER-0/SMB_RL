# trimming

Fairly aggressive trimming is used to try and remove extraneous information from the screen.

However, in edge cases, the player can access these regions. A fully developed model that learned to make use of the warp zone in course 1-2 for instance, would be hampered by this trimming. Below is an image where the player character is in the portion of the screen nominally reserved for the the score and other player information.

Nonetheless, we believe there is value to reducing the size of the input image for the neural network and given the hidden nature of this edge case, it seems reasonable to ignore.

That said, it is unclear how often upward segments (eg., vines) appear and would be negatively affected by this.
